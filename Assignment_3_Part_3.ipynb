```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#reading the dataset
train_data = pd.read_csv('train.csv')
train_data.head()
```

```python
#performing exploration of the dataset
train_data.info()
```

```python
train_data.describe()
```

```python
#Calculate the Pearson correlation matrix
correlation_matrix = train_data.corr()
correlation_matrix
```

```python
correlation = train_data['SalePrice'].corr(train_data['OverallQual'])
correlation
```

```python
var ='OverallQual'
data = pd.concat([train_data['SalePrice'], train_data[var]], axis =1)
f, ax = plt.subplots(figsize=(8,6))
fig = sns.boxplot(x=var, y="SalePrice", data=train_data)
fig.axis(ymin=0, ymax=800000);
```

```python
correlation = train_data['SalePrice'].corr(train_data['GrLivArea'])
correlation
```

```python
plt.scatter(train_data['SalePrice'], train_data['GrLivArea'])
plt.xlabel('SalePrice')
plt.ylabel('GrLivArea')
plt.title("Correlation between Sales Price and GrLivArea")
```
